```python
#To create a feature matrixs, Here x is independent variables and y is dependent variable/target variable
x= df.iloc[:,[4,5,6,7,8,9,10]].values
y= df.iloc[:,11].values
```

```python
x
```

```python
# Data spliting ,here 80% training 20% testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=0)
```

```python

x_train.shape
```

```python
from math import sqrt
sqrt(65014)
```

## RndomFoestRegressor Model

```python
from sklearn.ensemble import RandomForestRegressor
reg =RandomForestRegressor(n_estimators=255,random_state=0)
reg.fit(x_train,y_train)
```

```python
# prediction
y_pred = reg.predict(x_test)
```

```python
y_pred
```

```python
# r2_score 
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

### Evaluation Model

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy = cross_val_score(estimator=reg,X=x_train,y=y_train,cv=10)
```
